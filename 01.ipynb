```python
#Thêm thư viện
from PIL import Image
```

```python
#Khai báo một biến chứa thư mục
dir = 'C:/Thigiacmaytinh'
```

```python
#Khai báo đường dẫn ảnh
image_path = dir +'/4.jpg'
print(image_path)
```

```python
#Sử dụng Image.open() để đọc ảnh
img04 = Image.open(image_path)
```

```python
#Hiển thị ảnh 
display(img04)
```

```python
 #In thông tin cơ bản
print("Định dạng ảnh : ", img04.format)
```

```python
#In kích cỡ ảnh 
print("Kích cỡ ảnh : ", img04.size)
```

```python
#Lưu ảnh và chuyển đổi ảnh
new_img_04_path = dir + '/new_01.JPG'
new_img_png_04_path = dir + '/new_01.PNG'
img04.save(new_img_04_path)
img04.save(new_img_png_04_path)
```

```python
#Đóng tập tin
img04.close()
```
